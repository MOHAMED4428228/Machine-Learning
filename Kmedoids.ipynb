```python
from sklearn.preprocessing import StandardScaler

X = df.drop('class', axis=1)
y = df['class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("Original features (first 5 rows):")
display(X.head())
print("\nScaled features (first 5 rows):")
display(X_scaled_df.head())
```

```python
!pip install --upgrade pip
!pip install pyclustering

from pyclustering.cluster.kmedoids import kmedoids
from sklearn.metrics import silhouette_score
import numpy as np

inertia = []
silhouette_scores = []

k_range = range(2, 11)

X_scaled_list = X_scaled.tolist()

for k in k_range:

    initial_medoids = np.random.choice(X_scaled.shape[0], k, replace=False)

    instance_kmedoids = kmedoids(X_scaled_list, initial_medoids, random_state=42)

    instance_kmedoids.process()

    clusters = instance_kmedoids.get_clusters()
    medoids = instance_kmedoids.get_medoids()


    labels = np.zeros(X_scaled.shape[0], dtype=int)
    for cluster_idx, cluster in enumerate(clusters):
        for point_idx in cluster:
            labels[point_idx] = cluster_idx

    current_inertia = 0
    for cluster_idx, cluster_points_indices in enumerate(clusters):
        if len(cluster_points_indices) > 0:
            medoid_point = X_scaled[medoids[cluster_idx]]
            for point_idx in cluster_points_indices:
                current_inertia += np.sum((X_scaled[point_idx] - medoid_point)**2)
    inertia.append(current_inertia)

    if k > 1:
        score = silhouette_score(X_scaled, labels)
        silhouette_scores.append(score)
    else:
        silhouette_scores.append(0)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Sum of Squared Errors)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
optimal_k = 3

X_scaled_list = X_scaled.tolist()

initial_medoids = np.random.choice(X_scaled.shape[0], optimal_k, replace=False)

instance_kmedoids_optimal = kmedoids(X_scaled_list, initial_medoids, random_state=42)
instance_kmedoids_optimal.process()

clusters_optimal = instance_kmedoids_optimal.get_clusters()

optimal_labels = np.zeros(X_scaled.shape[0], dtype=int)
for cluster_idx, cluster in enumerate(clusters_optimal):
    for point_idx in cluster:
        optimal_labels[point_idx] = cluster_idx

df['KMedoids_Cluster'] = optimal_labels

print(f"DataFrame with K-Medoids cluster labels (first 5 rows) for k={optimal_k}:")
display(df.head())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

pca_df = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])

pca_df['KMedoids_Cluster'] = df['KMedoids_Cluster']

plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='Principal Component 1',
    y='Principal Component 2',
    hue='KMedoids_Cluster',
    palette='viridis',
    data=pca_df,
    s=100,
    alpha=0.8
)
plt.title(f'K-Medoids Clusters (k={optimal_k}) Visualized with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.legend(title='Cluster')
plt.show()
```
